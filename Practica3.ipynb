1-.Crear un Dataframe

```python
import pandas as pd
data = {'precio': [8000, 12000, 9500, 15000, 7000],
        'direccion': ['Calle A, 123', 'Avenida B, 45', 'Plaza C, 67', 'Ronda D, 89', 'Paseo E, 10']}
df = pd.DataFrame(data)
print("--- 1. DataFrame creado ---")
print(df)
print("\n")
```

2-.Dar toda la Informacion

```python
print("--- 2. Información del DataFrame ---")
print(df.info())
print("\n")
```

Crear una mascara y van a filtrar por precio >=10,000

```python
mascara = df['precio'] >= 10000
df_filtrado = df[mascara]
print("--- 3. Casas con precio >= 10,000 ---")
print(df_filtrado)
print("\n")
```

¿Cuantos son inferiores a 10,000?

```python
menores = df[~mascara]
cantidad = len(menores)
print("--- 4. Cantidad de casas con precio < 10,000 ---")
print(cantidad)
print("\n")
```

Crear un nuevo Dataframe

```python
nuevo_df = df_filtrado.copy()
print("--- 5. Nuevo DataFrame creado ---")
print(nuevo_df)
print("\n")
```

Dar toda la informacion

```python
print("--- 6. Información del nuevo DataFrame ---")
print(nuevo_df.info())
print("\n")
```

Crear una serie con todas las dirreciones de las casas >=10,000

```python
serie_direcciones = nuevo_df['direccion']
print("--- 7. Serie de direcciones (precio >= 10,000) ---")
print(type(serie_direcciones)) # Verificamos que es una Serie
print(serie_direcciones)
```
